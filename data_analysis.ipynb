```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('world_gdp.csv')
df2 = np.transpose(df)
df
```

```python
countries = {i:df2[i][0] for i in df2}
df2.drop('Country Name',inplace=True)
```

```python
yearly_change = {}
total_change = {}
percent_change = {}
for i in df2:
    yearly_change[i] = np.diff(df2[i])
    total_change[i] = df2[i][-1] - df2[i][0]
    percent_change[i] = (total_change[i]/df2[i][0])
```

```python
#Decreasing gdp
for i in total_change:
    if(total_change[i]<0):
        print(countries[i],percent_change[i]*100,total_change[i])
```

```python
#Increasing gdp (Value)
l = sorted(total_change,key = lambda x:total_change[x],reverse = True)
for i in l[:15]:
    print(countries[i],total_change[i])
```

```python
#Increasing gdp (Percentage)
l = sorted(percent_change,key = lambda x:percent_change[x],reverse = True)
for i in l[:10]:
    print(countries[i],percent_change[i])
```

```python
years = {i:0 for i in range(2000,2017)}
for i in yearly_change:
    temp = df2[i]
    for k,j in enumerate(yearly_change[i]):
        if(j>0):
            years[k+2000] += 1
sort_years = sorted(years,key = lambda x: years[x], reverse = True)
print("Good Years:")
for i in sort_years[:5]:
    print(i,years[i])
print("-------------------")
print("Bad Years:")
for i in sort_years[:-6:-1]:
    print(i,years[i])
print("-------------------")
```

```python
good_years = {}
bad_years = {}
for i in yearly_change:
    temp = df2[i]
    good_years[i] = []
    bad_years[i] = []
    for k,j in enumerate(yearly_change[i]):
        if(j>0):
            good_years[i].append(k+2000)
        if(j<0):
            bad_years[i].append(k+2000)
#India 82
#Pakistan 142
#United States 188
#China 34
for i in  bad_years:
    if(bad_years[i]==[]):
        print(countries[i])
print("-------------------")        
for i in [82,142,188,34]:
    print(countries[i])
    print("Good",good_years[i])
    print("Bad",bad_years[i])
    print('-----------------')
```
